```python

Trend_data = TrendReq(hl='en-US', tz=360)
Trend_data.build_payload(kw_list=['Data Analyst'])
Trend_data = Trend_data.interest_over_time()
fig, ax = plt.subplots(figsize=(19, 15))
Trend_data['Data Analyst'].plot()
plt.style.use('fivethirtyeight')
plt.title('Google Search Trend for Data Analyst', fontweight='bold')
plt.xlabel('Trends over Year')
plt.ylabel('Trend Data')
plt.show()
```
